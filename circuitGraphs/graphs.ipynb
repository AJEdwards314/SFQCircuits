Universal function

```python
import matplotlib.pyplot as plt

def graphCsv(inputFile):
    op = open(inputFile)
    data = op.readlines()
    colNames = data[0].split(',"')
    for i in range(len(colNames)):
        colNames[i] = colNames[i].replace('"','')
    data = data[1:]
    nums = []
    time=[]
    for i in range(len(data)):
        nums.append(data[i].split(','))
        for j in range(len(nums[i])):
            pH = float(nums[i][j])
            nums[i][j] = pH
            if j == 0:
                time.append(pH)
    for i in range(1,len(nums[0])):
        yAxis = []
        for j in range(len(nums)):
            yAxis.append(nums[j][i])
        plt.plot(time, yAxis, label=colNames[i])
        plt.title(colNames[i])
        plt.legend()
        plt.show()
    for i in range(1,len(nums[0])):
        yAxis = []
        for j in range(len(nums)):
            yAxis.append(nums[j][i])
        plt.plot(time, yAxis, label=colNames[i])
    plt.title("everything together (scale may make graph bad)")
    plt.legend()
    plt.show()
```
